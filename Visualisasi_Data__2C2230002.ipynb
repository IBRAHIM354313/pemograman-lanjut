1.Histogram distribusi nilai matematika.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Membaca file CSV
df = pd.read_csv('/content/H05.csv')
# Membuat histogram orang berdasarkan kelompok umur
df[["nilai_mat"]].plot(kind="hist", bins=[0, 20, 40, 60, 80, 100], rwidth=0.8, title="Histogram distribusi nilai matematika")
# Menampilkan plot
plt.show()
```

2.Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.

```python
plt.figure(figsize=(8, 5))
df['fakultas'].value_counts().plot(kind='barh', color="red", edgecolor="black")
plt.title('Jumlah Mahasiswa per Fakultas')
plt.xlabel('Jumlah Mahasiswa')
plt.ylabel('Fakultas')
plt.show()
```

3.Diagram pie banyaknya mahasiswa masing-masing fakultas

```python
plt.figure(figsize=(8, 8))
df['fakultas'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Presentase Mahasiswa per Fakultas')
plt.show()
```

5.Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['nilai_kim'], df['nilai_fis'])
plt.title('Hubungan Nilai Kimia dan Fisika')
plt.xlabel('Nilai Kimia')
plt.ylabel('Nilai Fisika')
plt.show()
```
